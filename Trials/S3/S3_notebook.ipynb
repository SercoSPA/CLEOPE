# Sentinel-3 data visualization

With this notebook you can easily unpack and visualize Sentinel-3 WFR and LFR products on a georeferenced plot. Products are taken from the dataframe imported from SEARCH notebook. 

```python
import data_processing as dp
```

Read the products dataframe, saving results in a product list.

```python
products = dp.products()
```

In the following lines we have splitted S3 products into two main functions: land and water. 

Land function will detect OLCI L2 LFR products, picking up a variable among OTCI (OLCI Terrestrial Clorophyll Index) and OGVI (OLCI Global Vegetation Index). On the other hand, water function will pick up OLCI L2 WFR products and supports the visualization of the TSM variable (the Total Suspended Matter in the sea), the variable CHL_OC4ME which indicates the clorophyll richness in the algal pigments and and IWV (Integrated Water Vapor column density). 

Once the variable of interest has been specified, land and water functions will read S3 .nc binary data, selecting the 2D matrixes of values and the relative coordinates files. In this way it is possible to show S3 OLCI products on a geo-referenced plot.

Choose for `var` one of the following:

| `land` | `water` |
| --- | --- |
| `otci` | `tsm_nn` |
| `ogvi` | `chl_oc4me` |

```python
i = 0
for p in products:
    if "LFR" in p.split("/")[-1]:
        filename = "otci_index"
        dp.land(p,filename+str(i),var="otci")
    elif "WFR" in p.split("/")[-1]:
        filename = "tsm"
        dp.water(p,filename+str(i),var="tsm_nn")
    i+=1
```
